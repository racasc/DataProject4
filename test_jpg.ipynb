```python
#Lectura de archivos dicom
def load_image(sample_path):
    ds = pydicom.dcmread(sample_path)
    img = ds.pixel_array
    p01 = np.percentile(img.ravel(), 1)
    p99 = np.percentile(img.ravel(), 99)
    img[img > p99] = p99
    img[img < p01] = p01
    if ds.data_element('PhotometricInterpretation').value == 'MONOCHROME1':
        img = np.invert(img)
    img = (img - img.min()) / (img.max() - img.min())
    return img
```

```python
#Creacion de archivos .jpg
import cv2
import os
import pydicom

inputdir = '/content/test/'
outdir = '/content/test-jpg/'


test_list = [ f for f in  os.listdir(inputdir)]

for f in test_list:
    ds = load_image(inputdir + f) # leer archivo dicom
    cv2.imwrite(outdir + f.replace('.dcm','.jpg'),(ds*255).astype('uint8')) # escritura de arhico .jpg
```
